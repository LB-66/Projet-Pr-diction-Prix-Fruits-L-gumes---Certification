## 0. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

ROUGE = '#C0392B'
BLEU  = '#3498DB'
VERT  = '#2ECC71'

print('OK')
```

## 1. Chargement

Le fichier `ers_toutes_annees.csv` a été compilé depuis les archives USDA ERS.
Il contient les prix moyens de 132 fruits et légumes sur 5 années (2013, 2016, 2020, 2022, 2023),
soit 704 lignes au total.

**Colonnes** :
- `produit` : nom du fruit ou légume
- `forme` : Fresh / Canned / Frozen / Juice / Dried
- `prix_detail` : prix moyen en supermarché (\$/lb)
- `rendement` : proportion utilisable après préparation (0 à 1)
- `taille_cup` : poids d'une portion standard en cup
- `prix_cup` : **variable cible** — coût réel d'une portion (\$/cup)
- `categorie` : fruit ou legume
- `annee` : année de collecte des prix

```python
df = pd.read_csv('ers_toutes_annees.csv', sep=';', encoding='utf-8')

print(f'Lignes    : {df.shape[0]}')
print(f'Colonnes  : {df.shape[1]}')
print(f'Annees    : {sorted(df["annee"].unique())}')
print(f'Produits  : {df["produit"].nunique()}')
df.head(5)
```

## 2. Exploration rapide

Avant de nettoyer, on regarde l'état brut du dataset.

```python
# Statistiques descriptives
df.describe()
```

```python
# Valeurs manquantes
print('Valeurs manquantes :')
print(df.isnull().sum())

# Repartition par annee
print('\nRepartition par annee et categorie :')
print(df.groupby(['annee', 'categorie']).size().unstack(fill_value=0))
```

## 3. Nettoyage

Le dataset est déjà propre (pas de valeurs manquantes, pas de doublons).
On applique trois traitements simples :
1. Suppression des espaces dans les colonnes texte
2. Encodage numérique de la forme et de la catégorie pour le ML
3. Vérification des types

```python
# Nettoyage des espaces
df['produit']   = df['produit'].str.strip()
df['forme']     = df['forme'].str.strip()
df['categorie'] = df['categorie'].str.strip()

# Encodage de la forme pour le modele ML
# Fresh=0, Canned=1, Frozen=2, Juice=3, Dried=4
forme_map = {'Fresh': 0, 'Canned': 1, 'Frozen': 2, 'Juice': 3, 'Dried': 4}
df['forme_encoded'] = df['forme'].map(forme_map)

# Encodage de la categorie : fruit=1, legume=0
df['categorie_encoded'] = (df['categorie'] == 'fruit').astype(int)

print('Types des colonnes :')
print(df.dtypes)
print(f'\nDataset apres nettoyage : {df.shape[0]} lignes x {df.shape[1]} colonnes')
```

## 4. Variable cible

On classe chaque ligne en 3 niveaux de prix par **terciles** :
- **Peu cher** : 33% les moins chers
- **Modere** : 33% du milieu
- **Cher** : 33% les plus chers

Cette variable cible sera utilisée pour la classification.
Pour la régression, on utilisera directement `prix_cup` comme valeur continue.

```python
t1 = df['prix_cup'].quantile(1/3)
t2 = df['prix_cup'].quantile(2/3)

df['niveau_prix'] = pd.cut(
    df['prix_cup'],
    bins=[-0.01, t1, t2, 999],
    labels=['Peu cher', 'Modere', 'Cher']
)

print('Repartition de la variable cible :')
print(df['niveau_prix'].value_counts())
print(f'\nSeuils :')
print(f'  Peu cher : prix_cup < {t1:.2f}$')
print(f'  Modere   : {t1:.2f}$ <= prix_cup < {t2:.2f}$')
print(f'  Cher     : prix_cup >= {t2:.2f}$')

print('\nExemples par classe :')
for c in ['Peu cher', 'Modere', 'Cher']:
    ex = df[df['niveau_prix'] == c]['produit'].unique()[:3]
    print(f'  {c:10} : {list(ex)}')
```

## 5. Visualisations

```python
# Variable cible
compte  = df['niveau_prix'].value_counts()
palette = {'Peu cher': VERT, 'Modere': BLEU, 'Cher': ROUGE}

plt.figure(figsize=(6, 4))
plt.bar(compte.index, compte.values,
        color=[palette[c] for c in compte.index])
plt.title('Repartition de la variable cible', fontweight='bold')
plt.ylabel('Nombre de lignes')
plt.show()
```

```python
# Evolution du prix moyen par annee
prix_annee = df.groupby(['annee', 'categorie'])['prix_cup'].mean().unstack()

plt.figure(figsize=(9, 5))
plt.plot(prix_annee.index, prix_annee['fruit'],
         color=ROUGE, marker='o', label='Fruits')
plt.plot(prix_annee.index, prix_annee['legume'],
         color=VERT, marker='o', label='Legumes')
plt.title('Evolution du prix moyen par cup (2013-2023)',
          fontweight='bold')
plt.xlabel('Annee')
plt.ylabel('Prix moyen par cup ($)')
plt.legend()
plt.xticks(prix_annee.index)
plt.tight_layout()
plt.show()
```

```python
# Prix moyen par forme
moy_forme = df.groupby('forme')['prix_cup'].mean().sort_values()

plt.figure(figsize=(7, 4))
plt.bar(moy_forme.index, moy_forme.values, color=BLEU)
plt.title('Prix moyen par cup selon la forme', fontweight='bold')
plt.ylabel('Prix moyen ($)')
for i, v in enumerate(moy_forme.values):
    plt.text(i, v + 0.01, f'{v:.2f}$', ha='center', fontsize=9)
plt.show()
```

```python
# Top 15 produits les plus chers
top = df.groupby('produit')['prix_cup'].mean().nlargest(15).reset_index()

plt.figure(figsize=(9, 6))
plt.barh(top['produit'], top['prix_cup'], color=ROUGE)
plt.title('Top 15 produits les plus chers par cup', fontweight='bold')
plt.xlabel('Prix moyen par cup ($)')
plt.tight_layout()
plt.show()
```

```python
# Top 15 produits les moins chers
flop = df.groupby('produit')['prix_cup'].mean().nsmallest(15).reset_index()

plt.figure(figsize=(9, 6))
plt.barh(flop['produit'], flop['prix_cup'], color=VERT)
plt.title('Top 15 produits les moins chers par cup', fontweight='bold')
plt.xlabel('Prix moyen par cup ($)')
plt.tight_layout()
plt.show()
```

```python
# Matrice de correlation
cols = ['prix_detail', 'rendement', 'taille_cup', 'prix_cup', 'annee']

plt.figure(figsize=(7, 5))
sns.heatmap(df[cols].corr(), annot=True, fmt='.2f',
            cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Matrice de correlation', fontweight='bold')
plt.tight_layout()
plt.show()
```
