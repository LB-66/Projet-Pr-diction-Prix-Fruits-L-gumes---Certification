## SOURCE 1 — API NASS : État de production

L'API NASS (USDA National Agricultural Statistics Service) donne
le volume de production par état pour chaque produit.
Cela permet de savoir où est cultivé chaque fruit ou légume.

```python
# Clé API NASS — inscription gratuite sur quickstats.nass.usda.gov
CLE_NASS = 'A9273C62-28A4-33B3-9E51-EFBD2EA33AED'

# Table de correspondance : nom du produit ERS → code NASS + unité
ERS_VERS_NASS = {
    'Apples'       : ('APPLES',       'LB'),
    'Blueberries'  : ('BLUEBERRIES',  'LB'),
    'Strawberries' : ('STRAWBERRIES', 'CWT'),
    'Tomatoes'     : ('TOMATOES',     'CWT'),
    'Broccoli'     : ('BROCCOLI',     'CWT'),
    'Grapes'       : ('GRAPES',       'TONS'),
    'Peaches'      : ('PEACHES',      'TONS'),
    'Cherries'     : ('CHERRIES',     'LB'),
    'Oranges'      : ('ORANGES',      'TONS'),
    'Carrots'      : ('CARROTS',      'CWT'),
    'Lettuce'      : ('LETTUCE',      'CWT'),
    'Potatoes'     : ('POTATOES',     'CWT'),
    'Spinach'      : ('SPINACH',      'CWT'),
    'Raspberries'  : ('RASPBERRIES',  'LB'),
    'Asparagus'    : ('ASPARAGUS',    'CWT'),
    'Onions'       : ('ONIONS',       'CWT'),
    'Corn'         : ('CORN',         'TONS'),
    'Cucumbers'    : ('CUCUMBERS',    'CWT'),
    'Cauliflower'  : ('CAULIFLOWER',  'CWT'),
    'Celery'       : ('CELERY',       'CWT'),
}

print(f'Produits dans la table NASS : {len(ERS_VERS_NASS)}')
```

```python
def appeler_nass(produit_nass, unite, cle):
    """Récupère la production par état pour un produit via l'API NASS."""
    url = 'https://quickstats.nass.usda.gov/api/api_GET/'
    params = {
        'key'               : cle,
        'commodity_desc'    : produit_nass,
        'statisticcat_desc' : 'PRODUCTION',
        'unit_desc'         : unite,
        'year'              : 2023,
        'agg_level_desc'    : 'STATE',
        'format'            : 'JSON'
    }
    try:
        time.sleep(0.3)
        rep = requests.get(url, params=params, timeout=15)
        if rep.status_code == 200:
            donnees = rep.json().get('data', [])
            print(f'  {produit_nass} : {len(donnees)} etats')
            return donnees
        return []
    except Exception as e:
        print(f'  {produit_nass} : erreur — {e}')
        return []


print('Appel API NASS...')
resultats_nass = []

# On déduplique pour éviter d'appeler deux fois le même produit NASS
produits_vus = {}
for nom_ers, (nom_nass, unite) in ERS_VERS_NASS.items():
    if nom_nass not in produits_vus:
        produits_vus[nom_nass] = []
    produits_vus[nom_nass].append(nom_ers)

for nom_nass, noms_ers in produits_vus.items():
    unite = ERS_VERS_NASS[noms_ers[0]][1]
    donnees = appeler_nass(nom_nass, unite, CLE_NASS)
    for ligne in donnees:
        # On filtre les lignes OTHER STATES non exploitables
        if ligne.get('state_name') == 'OTHER STATES':
            continue
        for nom_ers in noms_ers:
            resultats_nass.append({
                'produit'       : nom_ers,
                'etat'          : ligne.get('state_name'),
                'code_etat'     : ligne.get('state_alpha'),
                'production_raw': ligne.get('Value'),
                'unite'         : unite,
            })

df_nass = pd.DataFrame(resultats_nass)
print(f'\nNASS : {df_nass.shape[0]} lignes récupérées')
```

```python
# Nettoyage des données NASS
# Conversion de la production en nombre (suppression des virgules)
df_nass['production'] = pd.to_numeric(
    df_nass['production_raw'].astype(str).str.replace(',', ''),
    errors='coerce'
)

# Conversion de toutes les unités en livres (LB)
CONVERSION = {'LB': 1, 'CWT': 100, 'TONS': 2000, 'BU': 56}
df_nass['production_lbs'] = df_nass.apply(
    lambda r: r['production'] * CONVERSION.get(r['unite'], 1), axis=1
)

# Suppression des valeurs nulles ou zéro
df_nass = df_nass[df_nass['production_lbs'] > 0].dropna(subset=['production_lbs'])

# On garde uniquement l'état avec le plus gros volume par produit
# C'est cet état qu'on utilisera pour la météo
idx = df_nass.groupby('produit')['production_lbs'].idxmax()
df_etat = df_nass.loc[idx][['produit', 'etat', 'code_etat', 'production_lbs']].copy()
df_etat.columns = ['produit', 'etat_production', 'code_etat', 'production_lbs']
df_etat = df_etat.reset_index(drop=True)

print('État principal par produit :')
print(df_etat.to_string())
```

```python
# Jointure NASS avec le dataset principal
# On extrait le premier mot du nom du produit pour matcher
df['produit_cle'] = df['produit'].str.split(',').str[0].str.strip()

df = df.merge(
    df_etat,
    left_on='produit_cle',
    right_on='produit',
    how='left'
)

# Nettoyage des colonnes dupliquées après la jointure
df.drop(columns=['produit_y', 'produit_cle'], inplace=True, errors='ignore')
df.rename(columns={'produit_x': 'produit'}, inplace=True)

pct = df['etat_production'].notna().mean() * 100
print(f'Taux de correspondance NASS : {pct:.1f}%')
print(f'Dataset : {df.shape[0]} lignes x {df.shape[1]} colonnes')
```

## SOURCE 2 — API Open-Meteo : Météo historique

Open-Meteo fournit des données météo historiques gratuitement, sans clé API.
On récupère la météo 2023 pour chaque état producteur identifié via le NASS.

```python
# Coordonnées GPS des états producteurs
GPS_ETATS = {
    'CA': (36.7783, -119.4179),
    'WA': (47.3826, -120.4472),
    'FL': (27.9944,  -81.7603),
    'MI': (43.3504,  -84.5603),
    'OR': (44.5720, -122.0709),
    'GA': (32.1574,  -82.9071),
    'NC': (35.6301,  -79.8064),
    'ID': (44.0682, -114.7420),
    'IA': (42.0046,  -93.2140),
    'WI': (44.2685,  -89.6165),
    'NY': (42.1657,  -74.9481),
    'PA': (40.5908,  -77.2098),
    'MT': (46.8797, -110.3626),
    'HI': (20.7967, -156.3319),
    'TX': (31.9686,  -99.9018),
    'MN': (46.3296,  -94.3213),
    'NJ': (40.2989,  -74.5210),
}

# Ajout des coordonnées GPS dans la table des états
df_etat['latitude']  = df_etat['code_etat'].map(
    lambda x: GPS_ETATS.get(x, (None, None))[0])
df_etat['longitude'] = df_etat['code_etat'].map(
    lambda x: GPS_ETATS.get(x, (None, None))[1])

print('Coordonnées GPS ajoutées')
print(df_etat[['produit', 'etat_production', 'latitude', 'longitude']].head(10))
```

```python
def appeler_meteo(lat, lon, annee=2023):
    """Récupère les données météo annuelles pour des coordonnées GPS."""
    url = 'https://archive-api.open-meteo.com/v1/archive'
    params = {
        'latitude'  : lat,
        'longitude' : lon,
        'start_date': f'{annee}-01-01',
        'end_date'  : f'{annee}-12-31',
        'daily'     : 'temperature_2m_max,temperature_2m_min,precipitation_sum',
        'timezone'  : 'America/New_York'
    }
    try:
        time.sleep(0.5)
        rep = requests.get(url, params=params, timeout=20)
        if rep.status_code == 200:
            d     = rep.json().get('daily', {})
            tmax  = d.get('temperature_2m_max', [])
            tmin  = d.get('temperature_2m_min', [])
            prec  = d.get('precipitation_sum', [])
            # Calcul des indicateurs agrométéorologiques
            return {
                'temp_moyenne'   : round(np.nanmean([(a+b)/2 for a,b in zip(tmax,tmin)]), 2),
                'jours_gel'      : sum(1 for t in tmin if t is not None and t < 0),
                'jours_chaleur'  : sum(1 for t in tmax if t is not None and t > 35),
                'precip_totale'  : round(sum(p for p in prec if p is not None), 1),
            }
        return None
    except Exception as e:
        print(f'  Erreur météo : {e}')
        return None


print('Appel Open-Meteo...')
resultats_meteo = []

for _, row in df_etat.iterrows():
    if pd.isna(row['latitude']):
        continue
    meteo = appeler_meteo(row['latitude'], row['longitude'])
    if meteo:
        meteo['produit']         = row['produit']
        meteo['etat_production'] = row['etat_production']
        resultats_meteo.append(meteo)
        print(f'  {row["produit"]} ({row["etat_production"]}) OK')

df_meteo = pd.DataFrame(resultats_meteo)
print(f'\nMétéo récupérée : {df_meteo.shape[0]} produits')
```

```python
# Jointure météo avec le dataset principal
df['produit_cle'] = df['produit'].str.split(',').str[0].str.strip()

df = df.merge(
    df_meteo[['produit', 'temp_moyenne', 'jours_gel',
              'jours_chaleur', 'precip_totale']],
    left_on='produit_cle',
    right_on='produit',
    how='left'
)

df.drop(columns=['produit_y', 'produit_cle'], inplace=True, errors='ignore')
df.rename(columns={'produit_x': 'produit'}, inplace=True)

pct = df['temp_moyenne'].notna().mean() * 100
print(f'Taux de correspondance météo : {pct:.1f}%')
print(f'Dataset : {df.shape[0]} lignes x {df.shape[1]} colonnes')
```

## SOURCE 3 — API EIA : Prix de l'énergie

L'EIA publie les prix du diesel et de l'électricité par région américaine.
Ces prix impactent les coûts de transport et de stockage réfrigéré des produits.

```python
# Clé API EIA — inscription gratuite sur eia.gov/opendata
CLE_EIA = 'TA_CLE_EIA_ICI'

# Mapping état → région PADD (seules données diesel disponibles par EIA)
ETAT_VERS_PADD = {
    'CA': 'SCA', 'WA': 'R50', 'OR': 'R50', 'HI': 'R50',
    'ID': 'R40', 'MT': 'R40',
    'FL': 'R1Z', 'GA': 'R1Z', 'NC': 'R1Z', 'TX': 'R30',
    'PA': 'R1Y', 'NY': 'R1Y', 'NJ': 'R1Y',
    'MI': 'R20', 'IA': 'R20', 'WI': 'R20', 'MN': 'R20',
}

# États uniques dans notre dataset
ETATS_UNIQUES = df['code_etat'].dropna().unique()

def appeler_eia_diesel(padd, cle, annee=2023):
    """Récupère le prix moyen du diesel pour une région PADD."""
    url    = 'https://api.eia.gov/v2/petroleum/pri/gnd/data/'
    params = {
        'api_key'           : cle,
        'frequency'         : 'monthly',
        'data[0]'           : 'value',
        'facets[duoarea][]' : padd,
        'facets[product][]' : 'EPD2D',
        'facets[process][]' : 'PTE',
        'start'             : f'{annee}-01',
        'end'               : f'{annee}-12',
    }
    try:
        time.sleep(0.3)
        rep = requests.get(url, params=params, timeout=15)
        if rep.status_code == 200:
            donnees = rep.json().get('response', {}).get('data', [])
            valeurs = [float(d['value']) for d in donnees
                       if d.get('value') not in [None, 'null', '']]
            if valeurs:
                return round(np.mean(valeurs), 4)
        return None
    except Exception:
        return None

def appeler_eia_electricite(code_etat, cle, annee=2023):
    """Récupère le prix moyen de l'électricité pour un état."""
    url    = 'https://api.eia.gov/v2/electricity/retail-sales/data/'
    params = {
        'api_key'              : cle,
        'frequency'            : 'annual',
        'data[0]'              : 'price',
        'facets[stateid][]'    : code_etat,
        'facets[sectorid][]'   : 'ALL',
        'start'                : str(annee),
        'end'                  : str(annee),
    }
    try:
        time.sleep(0.3)
        rep = requests.get(url, params=params, timeout=15)
        if rep.status_code == 200:
            donnees = rep.json().get('response', {}).get('data', [])
            valeurs = [float(d['price']) for d in donnees
                       if d.get('price') not in [None, 'null', '']]
            if valeurs:
                return round(np.mean(valeurs), 4)
        return None
    except Exception:
        return None


print('Appel API EIA...')
resultats_eia = []

# On récupère les prix pour chaque état unique du dataset
for code in ETATS_UNIQUES:
    padd         = ETAT_VERS_PADD.get(code, 'NUS')
    prix_diesel  = appeler_eia_diesel(padd, CLE_EIA)
    prix_elec    = appeler_eia_electricite(code, CLE_EIA)
    resultats_eia.append({
        'code_etat'       : code,
        'prix_diesel'     : prix_diesel,
        'prix_electricite': prix_elec,
    })
    print(f'  {code} — diesel={prix_diesel}  elec={prix_elec}')

df_eia = pd.DataFrame(resultats_eia)

# Jointure EIA avec le dataset principal
df = df.merge(df_eia, on='code_etat', how='left')

# Remplacement des NaN par la médiane nationale
for col in ['prix_diesel', 'prix_electricite']:
    df[col].fillna(df[col].median(), inplace=True)

print(f'\nDataset : {df.shape[0]} lignes x {df.shape[1]} colonnes')
```

## SOURCE 4 — World Bank : Prix des engrais

Le World Bank Pink Sheet publie les prix mondiaux des engrais.
Ces prix impactent les coûts de production agricole avec un décalage de 6 mois.
Source : World Bank Pink Sheet Mars 2026

```python
# Prix des engrais 2023 — moyennes annuelles en $/mt
# Source : World Bank Pink Sheet Mars 2026
# Les prix sont identiques pour tous les produits car ce sont des prix mondiaux
ENGRAIS_2023 = {
    'urea_2023' : 358.0,
    'dap_2023'  : 550.0,
    'tsp_2023'  : 480.2,
    'mop_2023'  : 383.2,
}

# Prix 2022 avec décalage 6 mois — impact sur les prix alimentaires 2023
# En 2022 les prix étaient très élevés à cause de la crise Ukraine
ENGRAIS_LAG6M = {
    'urea_lag6m' : 520.0,
    'dap_lag6m'  : 780.0,
    'tsp_lag6m'  : 680.0,
    'mop_lag6m'  : 490.0,
}

# Ajout direct des colonnes engrais dans le dataset
for col, val in {**ENGRAIS_2023, **ENGRAIS_LAG6M}.items():
    df[col] = val

print('Colonnes engrais ajoutées :')
print(list(ENGRAIS_2023.keys()) + list(ENGRAIS_LAG6M.keys()))
print(f'Dataset : {df.shape[0]} lignes x {df.shape[1]} colonnes')
```

## SOURCE 5 — Scraping BLS : Prix retail

Le Bureau of Labor Statistics publie des prix moyens de détail
pour une sélection de fruits et légumes.
On scrape le tableau HTML depuis leur site web avec BeautifulSoup.

```python
# URL du tableau des prix BLS
URL_BLS = 'https://www.bls.gov/regions/mid-atlantic/data/averageretailfoodandenergyprices_usandwest_table.htm'
headers = {'User-Agent': 'Mozilla/5.0 (compatible; OrientIA-Bot/1.0)'}

print('Scraping BLS...')
df_bls = None

try:
    time.sleep(1)
    rep    = requests.get(URL_BLS, headers=headers, timeout=20)
    tables = pd.read_html(rep.text)
    table  = tables[0].copy()

    # Aplatissement des colonnes multi-niveaux
    table.columns = ['_'.join([str(c) for c in col
                     if 'Unnamed' not in str(c)]).strip()
                     for col in table.columns]

    # Extraction des colonnes utiles
    df_bls_raw = pd.DataFrame({
        'produit_bls': table.iloc[2:, 0].values,
        'prix_bls'   : table.iloc[2:, 2].values,
    })

    # Conversion en numérique
    df_bls_raw['prix_bls'] = pd.to_numeric(df_bls_raw['prix_bls'], errors='coerce')
    df_bls_raw.dropna(subset=['prix_bls'], inplace=True)
    df_bls_raw['produit_bls'] = df_bls_raw['produit_bls'].astype(str).str.strip()

    # Filtrage fruits et légumes uniquement
    mots = ['apple','banana','orange','grape','peach','strawberr',
            'potato','lettuce','tomato','broccoli','carrot',
            'celery','onion','corn','spinach','pepper']
    masque = df_bls_raw['produit_bls'].str.lower().str.contains(
        '|'.join(mots), na=False)
    df_bls = df_bls_raw[masque].copy()

    # Clé de jointure — premier mot du nom
    df_bls['produit_cle'] = df_bls['produit_bls'].str.split(',').str[0]\
                              .str.lower().str.strip()

    print(f'BLS : {len(df_bls)} produits scrapés')
    print(df_bls[['produit_bls', 'prix_bls']].to_string())

except Exception as e:
    print(f'Scraping échoué : {e}')
    df_bls = pd.DataFrame(columns=['produit_cle', 'prix_bls'])
```

```python
# Jointure BLS avec le dataset principal
df['produit_cle'] = df['produit'].str.split(',').str[0].str.lower().str.strip()

df = df.merge(
    df_bls[['produit_cle', 'prix_bls']],
    on='produit_cle',
    how='left'
)

# Remplacement des NaN par la médiane
df['prix_bls'].fillna(df['prix_bls'].median(), inplace=True)

# Suppression de la colonne temporaire de jointure
df.drop(columns=['produit_cle'], inplace=True, errors='ignore')

print(f'Dataset final : {df.shape[0]} lignes x {df.shape[1]} colonnes')
```
